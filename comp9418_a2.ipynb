```python
def sensor_reliability(data, sensors):
    
    sensors_reliability = []
    number_of_data = len(data)
    for sensor, area in sensors.items():
        # True positive: Sensor detects motion and the room has >=1 person
        # True negative: Sensor detects no motion and the room has no people
        data['tp'] = data.apply(lambda x: x[sensor] == 1 and x[area] > 0, axis=1)
        data['tn'] = data.apply(lambda x: x[sensor] == 0 and x[area] == 0, axis=1)
        
        prob_t = (data['tp'].sum() + data['tn'].sum()) / number_of_data
        prob_f = 1 - prob_t
        
        reliability = {}
        reliability[sensor] = (prob_t, prob_f)
        sensors_reliability.append(reliability)
        
    return sensors_reliability
```

```python
motion_sensors_reliability = sensor_reliability(df1, motion_sensors)
camera_sensors_reliability = sensor_reliability(df1, camera_sensors)
```

```python
# Disable the SettingWithCopyWarning
pd.options.mode.chained_assignment = None

def get_transition_matrix(data, room):
    
    room_data = data[['time', room]]
    # Set 1 if there's >=1 person in the room; 0 othersie
    room_data.loc[room_data[room] > 0, room] = 1
    room_data.loc[room_data[room] == 0, room] = 0
    
    # x_prev: Previous state (X_t-1) -> state of the previous timestamp 
    # x_curr: Current state (X_t) -> state of the current timestamp
    # State: If there're people in the room, then the light is on
    # Otherwise, the light is off
    room_data['x_prev'] = room_data[room].shift(1)
    room_data = room_data.rename(columns={room: 'x_curr'})
    room_data.drop(room_data.index[0], inplace=True)
    room_data['x_prev'] = room_data['x_prev'].astype('int64')
    room_data['x_curr'] = room_data['x_curr'].astype('int64')
    
    # Previous state = on
    prev_on = room_data.loc[(room_data['x_prev'] == 1)]
    # Previous state = off
    prev_off = room_data.loc[(room_data['x_prev'] == 0)]
    
    # Current state = on | previous state = on
    on_on = len(prev_on.loc[(prev_on['x_curr'] == 1)])
    prob_on_on = on_on / len(prev_on)
    # Current state = off | previous state = on
    on_off = len(prev_off.loc[(prev_off['x_curr'] == 1)])
    prob_on_off = on_off / len(prev_off)
    # Current state = on | previous state = off
    off_on = len(prev_on.loc[(prev_on['x_curr'] == 0)])
    prob_off_on = off_on / len(prev_on)
    # Current state = off | previous state = off
    off_off = len(prev_off.loc[(prev_off['x_curr'] == 0)]) 
    prob_off_off = off_off / len(prev_off)
    
    matrix = [[prob_on_on, prob_off_on], 
              [prob_on_off, prob_off_off]]
    return matrix
```

```python
rooms = ['r1', 'r2', 'r3', 'r4', 'r5', 'r6', 'r7', 'r8', 'r9', 'r10']
transition_matrices = []
for room in rooms: 
    room_matrix = get_transition_matrix(df1, room)
    transition_matrices.append(room_matrix)
    print(room)
    print(f'P(on|on): {room_matrix[0][0]}')
    print(f'P(off|on): {room_matrix[0][1]}')
    print(f'P(on|off): {room_matrix[1][0]}')
    print(f'P(off|off): {room_matrix[1][1]}')
    print('\n')
```
